```python
import matplotlib.pyplot as plt

#Sample data for energy consumption over 6 months (in Mich)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption= [1200, 1300, 1100, 1500, 1400, 1600]


#Create a Line plot
plt.plot(months, energy_consumption, marker='o', color='b', linestyle='--')

#Add titles and Labels
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Month')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption= [1200, 1300, 1100, 1500, 1400, 1600]

plt.pie(energy_consumption, labels=months, autopct='%1.1f%%', startangle=90)

plt.title('Energy Consumption Distribution Over 6 Months')

plt.show()
```

```python
 import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]
carbon_emissions = [400, 500, 450, 300, 350, 550]

plt.scatter(energy_consumption, carbon_emissions, marker='o', color='r', s=50)

plt.title('Energy Consumption vs. Carbon Emissions')
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emissions')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
"Solar": [1200, 1500, 1300],
"Wind": [3400, 3600, 3200],
"Hydropower": [2900, 3100, 2800],
"Blomass": [2500, 2700, 2400]
}

df =pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show('Correaltion Between Energy Sources')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Distrubution of Energy Consumption by Source')
plt.show()
```

```python
sns.violinplot(data=df)
plt.title('Violin plot of Energy Consumption Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

energy_values = [100, 200, 300, 400, 500]
carbon_emissions = [10, 20, 25, 40, 50]

df_reg = pd.DataFrame({'Energy': energy_values, 'Carbon Emissions': carbon_emissions})

sns.regplot(x='Energy', y='Carbon Emissions', data=df_reg)

plt.title('Relationship between Energy Consumption and Carbon Emissions')
plt
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

ax = plt.axes(projection='3d')

z= np.linspace(0,5,1000)
x= np.sin(25*z)
y= np.cos(25*z)

ax.plot3D(x,y,z, 'gray')
ax.set_title('3D Line Plot')
plt.show()
```
